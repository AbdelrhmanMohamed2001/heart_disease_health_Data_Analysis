```python
#DA=>has Disease and Drink Alcohol
#DNA=>has Disease and Not Drink Alcohol
#NDA=>hasn't Disease and Drink Alcohol
#NDNA=>hasn't Disease and Not Drink Alcohol
def AlcoholDrinking():

  DACholCheck=0
  DNACholCheck=0
  NDACholCheck=0
  NDNACholCheck=0
  DAHighChol=0
  DNAHighChol=0
  NDAHighChol=0
  NDNAHighChol=0
  DAHvyAlcoholConsump=0
  DNAHvyAlcoholConsump=0
  NDAHvyAlcoholConsump=0
  NDNAHvyAlcoholConsump=0

  for Disease, Alcohol in zip(new_df['HeartDiseaseorAttack'], new_df['CholCheck']):

      if Disease == 1 and Alcohol == 1:
          DACholCheck += 1
      elif Disease == 1 and Alcohol == 0:
          DNACholCheck += 1
      elif Disease ==0 and Alcohol == 1:
          NDACholCheck += 1
      elif Disease ==0 and Alcohol == 0:
          NDNACholCheck += 1
  TotalCholCheck= DACholCheck +DNACholCheck +NDACholCheck +NDNACholCheck
  print("Based On CholCheck Feature: ")
  print('Has Disease and Drink Alcohol ', DACholCheck, "Which Means ",(DACholCheck/TotalCholCheck)*100," % From Dataset")
  print("Has Disease and Not Drink Alcohol with CholCheck feature", DNACholCheck, "Which Means ",(DNACholCheck/TotalCholCheck)*100," % From Dataset")
  print("Hasn't Disease and Drink Alcohol with CholCheck feature", NDACholCheck, "Which Means ",(NDACholCheck/TotalCholCheck)*100," % From Dataset")
  print("Hasn't Disease and Not Drink Alcohol with CholCheck feature", NDNACholCheck, "Which Means ",(NDNACholCheck/TotalCholCheck)*100," % From Dataset")
  print(TotalCholCheck)
  print("*"*100)

  for Disease, Alcohol in zip(new_df['HeartDiseaseorAttack'], new_df['HighChol']):

      if Disease == 1 and Alcohol == 1:
          DAHighChol += 1
      elif Disease == 1 and Alcohol == 0:
          DNAHighChol += 1
      elif Disease ==0 and Alcohol == 1:
          NDAHighChol += 1
      elif Disease ==0 and Alcohol == 0:
          NDNAHighChol += 1
  TotalHighChol= DAHighChol+ DNAHighChol+ NDAHighChol+ NDNAHighChol
  print("Based On HighChol Feature: ")
  print('Has Disease and Drink Alcohol ', DAHighChol, "Which Means ",(DAHighChol/TotalHighChol)*100," % From Dataset")
  print("Has Disease and Not Drink Alcohol with CholCheck feature", DNAHighChol, "Which Means ",(DNAHighChol/TotalHighChol)*100," % From Dataset")
  print("Hasn't Disease and Drink Alcohol with CholCheck feature", NDAHighChol, "Which Means ",(NDAHighChol/TotalHighChol)*100," % From Dataset")
  print("Hasn't Disease and Not Drink Alcohol with CholCheck feature", NDNAHighChol, "Which Means ",(NDNAHighChol/TotalHighChol)*100," % From Dataset")
  print(TotalHighChol)


  print("*"*100)
  for Disease, Alcohol in zip(new_df['HeartDiseaseorAttack'], new_df['HvyAlcoholConsump']):

      if Disease == 1 and Alcohol == 1:
          DAHvyAlcoholConsump += 1
      elif Disease == 1 and Alcohol == 0:
          DNAHvyAlcoholConsump += 1
      elif Disease ==0 and Alcohol == 1:
          NDAHvyAlcoholConsump += 1
      elif Disease ==0 and Alcohol == 0:
          NDNAHvyAlcoholConsump += 1
  TotalHvyAlcoholConsump= DAHvyAlcoholConsump+ DNAHvyAlcoholConsump+ NDAHvyAlcoholConsump+ NDNAHvyAlcoholConsump
  print("Based On HvyAlcoholConsump Feature: ")
  print('Has Disease and Drink Alcohol ', DAHvyAlcoholConsump, "Which Means ",(DAHvyAlcoholConsump/TotalHvyAlcoholConsump)*100," % From Dataset")
  print("Has Disease and Not Drink Alcohol with CholCheck feature", DNAHvyAlcoholConsump, "Which Means ",(DNAHvyAlcoholConsump/TotalHvyAlcoholConsump)*100," % From Dataset")
  print("Hasn't Disease and Drink Alcohol with CholCheck feature", NDAHvyAlcoholConsump,"Which Means ",(NDAHvyAlcoholConsump/TotalHvyAlcoholConsump)*100," % From Dataset")
  print("Hasn't Disease and Not Drink Alcohol with CholCheck feature", NDNAHvyAlcoholConsump,"Which Means ",(NDNAHvyAlcoholConsump/TotalHvyAlcoholConsump)*100," % From Dataset")
  print(TotalHvyAlcoholConsump)





AlcoholDrinking()
```

**Relationship between HeartDiseaseorAttack and Smoking**

```python
#SD => smoke and has Disease
#SND =>smoke and hasn't Disease
#NSD=>Not smoke and has Disease
#NSND=>Not smoke and hasn't Disease

def Smoking():
  SD = 0
  SND = 0
  NSD=0
  NSND=0
  for Disease, smoke in zip(new_df['HeartDiseaseorAttack'], new_df['Smoker']):
     if Disease ==1 and smoke ==1 :
         SD += 1
     elif Disease ==0 and smoke ==1 :
         SND += 1
     elif Disease ==1 and smoke ==0 :
         NSD += 1
     elif Disease ==0 and smoke ==0 :
         NSND += 1

  Total= SD+ SND+ NSD+ NSND
  print(Total)
  print('smoke and has Disease ', SD, "Which Means ",(SD/Total)*100," % From Dataset")
  print("smoke and hasn't Disease ", SND, "Which Means ",(SND/Total)*100," % From Dataset")
  print("Not smoke and has Disease ", NSD, "Which Means ",(NSD/Total)*100," % From Dataset")
  print("Not smoke and hasn't Disease ", NSND, "Which Means ",(NSND/Total)*100," % From Dataset")
Smoking()
```

**Relationship between HeartDiseaseorAttack and Sex/Gender**

```python
#MD=> Male and has Disease
#FD=> Female and has Disease


gender_counts = df['Sex'].value_counts()
print("Number of Males:", gender_counts[1])
print("Number of Females:", gender_counts[0])



def gender():
    MD = 0
    FD = 0
    for Disease, gender in zip(new_df['HeartDiseaseorAttack'], new_df['Sex']):
        if Disease == 1 and gender == 1:
            MD += 1
        elif Disease == 1 and gender == 0:
            FD += 1


    print('Number of male that has Heart Disease', MD)
    print("*"*100)
    print('Number of female that has Heart Disease', FD)
    print("*"*100)
    print("Number of male that hasn't Heart Disease", gender_counts[1]-MD)
    print("*"*100)
    print("Number of female that hasn't Heart Disease", gender_counts[0]-FD)
    print("*"*100)

gender()
```

**Plot Features**

```python
def drawPlots():
  for column in new_df:
    plt.style.use('Solarize_Light2')
    plt.figure(figsize=(6, 8))
    plt.title(column)
    sns.countplot(x=column, data=new_df)
    plt.show()
drawPlots()
```

**Get Features Importance by Mutual Information of Dataset**

```python
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# X is your feature matrix and y is your target variable
mi_scores = mutual_info_classif(new_df, new_df['HeartDiseaseorAttack'])

# Print the mutual information scores for each feature
for feature, score in zip(new_df.columns, mi_scores):
    print(feature, score)
```

**Plot Mutual Information**

```python
import matplotlib.pyplot as plt

# Create a horizontal bar chart of the mutual information scores
fig, ax = plt.subplots()
y_pos = np.arange(len(new_df.columns))
ax.barh(y_pos, mi_scores, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(new_df.columns)
ax.set_xlabel('Mutual Information Score')
ax.set_title('Feature Importance')
plt.show()
```
